```python
# reload the model parameters
best_model = YOLO('../dataset_for_all/puncture_network/best_model/best.pt')
```

```python
metrics = best_model.val()
```

```python
test_results = best_model.val(data="../dataset_for_all/puncture_network/data.yaml", split='test')
```

```python
# validation accuracy
correct_prediction_count = 0
i = 0
path = "../dataset_for_all/puncture_network/valid/images"
dir_list = os.listdir(path)
for image_path in dir_list:
    file = open("../dataset_for_all/puncture_network/valid/labels/" + image_path[:-3] + "txt", "r")
    content = int(file.read()[0])
    # read/process image
    img_path = os.path.join("../dataset_for_all/puncture_network/valid/images", image_path)
    # image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
    inference_result = best_model(img_path)
    if (len(inference_result[0].boxes.data) == 0):
        if content == 0:
            correct_prediction_count += 1
        continue
    prob1 = inference_result[0].boxes.data[0][4].item()
    prob2 = inference_result[0].boxes.data[0][5].item()
    print("i = ", i)
    print("prob1 = ", prob1)
    print("prob2 = ", prob2)
    if (prob1 >= prob2 and content == 0):
        correct_prediction_count += 1
    elif (prob1 < prob2 and content == 1):
        correct_prediction_count += 1
    i += 1
print("correct_prediction_count = ", correct_prediction_count)
print("length of dir_list = ", len(dir_list))
print("success rate = ", correct_prediction_count / len(dir_list))
```
